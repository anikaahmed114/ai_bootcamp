```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('clipart4235836.png')
cv2_imshow(img)
```

```python
box_kernel = np.array([[1, 1, 1],
                      [1, 1, 1],
                      [1, 1, 1]]) / 9

gaussian_kernel = np.array([[1, 0, 0],
                            [0, 1, 0],
                            [0, 0, 1]]) / 9
```

```python
blurred_img = cv2.filter2D(img, -1, gaussian_kernel)

for i in range(3000):
  blurred_img = cv2.filter2D(blurred_img, -1, gaussian_kernel)

cv2_imshow(blurred_img)
```

Sobel's edge detection filter

```python
x_kernel = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]])

x_edges = cv2.filter2D(img, -1, x_kernel)
cv2_imshow(x_edges)
```

```python
y_kernel = np.array([[-1, -2, -1],
                     [0, 0, 0],
                     [1, 2, 1]])

y_edges = cv2.filter2D(img, -1, y_kernel)
cv2_imshow(y_edges)
```

```python
cv2_imshow(x_edges + y_edges)
```

Emboss Kernel

```python
emboss_kernel = np.array([[-2, -1, 1],
                     [-1, 1, 1],
                     [0, 1, 2]])

altered_img = cv2.filter2D(img, -1, emboss_kernel)
cv2_imshow(altered_img)
```
